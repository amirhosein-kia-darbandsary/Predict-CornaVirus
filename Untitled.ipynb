```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from tensorflow.keras.callbacks import EarlyStopping , ReduceLROnPlateau

from sklearn.utils import compute_class_weight
from tensorflow.keras.utils import plot_model
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
train_data_generated = ImageDataGenerator(
    rotation_range = 90 , 
    shear_range = 0.1 , 
    zoom_range=0.5 , 
    rescale=1./255.
)

test_data_generated = ImageDataGenerator(
    rescale=1. / 255.
)
```

```python
train = train_data_generated.flow_from_directory(
    './data' , 
    target_size=(500 , 500 ) , 
    batch_size=50 , 
    class_mode='binary',
    color_mode='grayscale'
)
```

```python
test = test_data_generated.flow_from_directory(
    'test_data' , 
    target_size=(500 , 500)  ,
    batch_size=50 , 
    class_mode='binary',
    color_mode='grayscale'
)
```

```python
model = tf.keras.Sequential(
[
    layers.Conv2D(64 , 3 , activation  = 'relu') , 
    layers.MaxPool2D(2), 
    layers.Conv2D(32 , 3  , activation = 'relu') , 
    layers.MaxPool2D(2) , 
    layers.Flatten(),
    layers.Dense(50 , activation = 'relu') , 
    layers.Dense(1 , activation = 'sigmoid')
    
])
```

```python
model.compile(
    optimizer= 'adam' , 
    loss = 'binary_crossentropy', 
    metrics=['accuracy']
)
```

```python
wh = compute_class_weight('balanced' , 
                     np.unique(train.classes) , 
                     train.classes)
```

```python
weights = dict(zip(np.unique(train.classes) , wh))
```

```python
er = EarlyStopping( monitor = 'val_accuracy' ,
                    patience = 10 , 
                    mode = 'max'
)

chm = ModelCheckpoint( filepath='./Modles/modle.h5' , 
                       monitor='val_accuracy' , 
                       mode = 'max' , 
                       save_best_only=True
                     )
```

```python
model.fit_generator(
    train , 
    steps_per_epoch=len(train) , 
    epochs=20 , 
    validation_data=test , 
    validation_steps=len(test) ,
    class_weight=weights , 
    callbacks = [er , chm]
    )
```
